```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Masood\Downloads\AB_NYC_2019.csv")
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
#dropping duplicates
```

```python
df=df.drop_duplicates()
df=df.drop_duplicates(subset=['id'],keep='first')
```

```python
#removing white spaces
```

```python
str_cols=df.select_dtypes(include=['object']).columns
for c in str_cols:
    df[c]=df[c].astype(str).str.strip().replace({'nan': pd.NA})
```

```python
#filling the host name and hostname with no name and missing host values
```

```python
df['name']=df['name'].fillna("No Name").replace({'<NA>': "No Name"})
df['host_name']=df['host_name'].fillna("Missing Host").replace({'<NA>': "Missing Host"})
```

```python
#coverting last review into date and filling the values with no data
```

```python
df['last_review']=pd.to_datetime(df['last_review'],errors='coerce')
```

```python
df['last_review']=df['last_review'].fillna("No Data").replace({'Nan': "No Data"})
```

```python
#filling reviews per month
```

```python
median=df['reviews_per_month'].median()
df['reviews_per_month']=df['reviews_per_month'].fillna(median).replace({'Nan':median})
```
